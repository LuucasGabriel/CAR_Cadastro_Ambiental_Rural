## Extraindo dados 

```python
#extraindo os dados somente dos municípios de interesse
import pandas as pd

df = pd.read_csv("caminho_para_seu_arquivo.csv")
municipios = [ "INÚBIA PAULISTA", "BASTOS", "TUPÃ", "HERCULÂNDIA", "SAGRES",
"OSVALDO CRUZ", "SALMOURÃO", "ARCO-ÍRIS", "LUCÉLIA", "QUEIROZ",
"IACRI", "PARAPUÃ", "RINÓPOLIS", "PRACINHA"]
df_sp = df[(df["uf"] == "SP") & (df["municipio"].isin(municipios))]
df_sp.to_csv("sobreposisao.csv", index=False)
```

## Formantando casas decimais

```python
import pandas as pd

# Carregar base
df = pd.read_csv("dados.csv")

# Corrigir formatação
df["area"] = (
    df["area"]
    .astype(str)
    .str.replace(".", "", regex=False)  # remove pontos de milhar
    .str.replace(",", ".", regex=False) # transforma vírgula em ponto decimal
    .astype(float)
)

df["area_formatada"] = df["area"].map(lambda x: f"{x:,.4f}".replace(",", "X").replace(".", ",").replace("X", "."))
df.to_csv("dados_corrigidos.csv", index=False, encoding="utf-8")
```
